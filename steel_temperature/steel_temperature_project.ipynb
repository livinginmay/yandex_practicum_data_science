Посчитаю сводные показатели по каждой партии:

```python
data_arc_sum = data_arc.groupby('key') \
    .agg({'heat_time': ['count', 'sum'], 'work': ['sum'], 'full_power': ['sum'], 'active_power':['sum'], 'reactive_power':['sum'], 'arc_start':['min'],'arc_end':['max']}).reset_index()
data_arc_sum.columns = ['key', 'heat_count', 'sum_heat_time', 'sum_work', 'sum_full_power', 'sum_active_power', 'sum_reactive_power', 'min_arc_start', 'max_arc_end']

data_arc_sum.head(10)
```

```python
data_arc_sum['all_time'] = (data_arc_sum['max_arc_end'] - data_arc_sum['min_arc_start']).dt.seconds
data_arc_sum['pause_time'] = data_arc_sum['all_time'] - data_arc_sum['sum_heat_time']
data_arc_sum.drop(['min_arc_start', 'max_arc_end'], axis=1, inplace=True)
data_arc_sum.head(10)
```

```python
data_arc_sum.describe()
```

```python
plots_on_columns(data_arc_sum, ['heat_count', 'sum_heat_time', 'sum_work', 'sum_full_power', 'sum_active_power', 'sum_reactive_power', 'all_time', 'pause_time'])
```

Добавлю признаки отдельно по каждой итерации нагрева. Пока это выглядит избыточно, но я буду смотреть на корреляцию и результаты метрик.

```python
#в LightSQL не нашла функцию ROW_NUMBER()
q =                                                    \
  "SELECT                                              \
    data_arc_1.*,                                      \
    COUNT(data_arc_2.key) AS iter_number               \
  FROM data_arc as data_arc_1                          \
    JOIN data_arc AS data_arc_2                        \
    ON data_arc_1.key = data_arc_2.key                 \
      AND data_arc_1.arc_start >= data_arc_2.arc_start \
  GROUP BY                                             \
    data_arc_1.key,                                    \
    data_arc_1.arc_start                               \
  ORDER BY                                             \
    data_arc_1.key,                                    \
    iter_number"
data_arc_iter = sqldf(q, globals())
data_arc_iter.head(10)
```

```python
data_frames = []
rename_columns = ['active_power', 'reactive_power', 'heat_time', 'full_power', 'work']
for i in range(1, 17):
  
    columns={}
    for j in range(0, 5):
        columns[rename_columns[j]] = rename_columns[j] + '_' + str(i)

    df = data_arc_iter.query('iter_number==@i').drop(['iter_number', 'arc_start', 'arc_end'], axis=1)
    df.rename(columns=columns, inplace=True)
    data_frames.append(df)

data_arc_full = reduce(lambda  left,right: pd.merge(left,right,on= ['key'], how='left'), data_frames)
data_arc_full.head()
```

```python
data_arc_full.columns
```

```python
data_arc_full = data_arc_sum.merge(data_arc_full, on = ['key'], how = 'inner')
data_arc_full.head(10)
```

```python
metrics = pd.DataFrame(columns = ['features_count', 'model_type', 'depth', 'n_estimators', 'mae', 'best_model'])
```

```python
def get_best_params(model, features_train, target_train, distributions=None, verbose=None):
    
    if distributions==None:
        distributions = {'max_depth':range(MIN_DEPTH, MAX_DEPTH + 1, DEPTH_STEP), 
                         'n_estimators':range(MIN_EST,MAX_EST + 1, EST_STEP)}  
      
    grid = RandomizedSearchCV(model, distributions, cv = CV, scoring = 'neg_mean_absolute_error', n_iter=N_ITER, 
                              n_jobs = -1, random_state=RANDOM_STATE)
    
    if verbose==None:
        grid.fit(features_train, target_train)
    else:
        grid.fit(features_train, target_train, verbose=verbose)
    
    return grid.best_estimator_, round(-grid.best_score_, 3)
```
